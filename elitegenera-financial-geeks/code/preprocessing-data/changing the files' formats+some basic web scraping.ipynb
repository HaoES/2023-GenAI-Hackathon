```python
import pandas as pd
from bs4 import BeautifulSoup
import os

file_path = r'C:\Users\Admin\Desktop\MoroccoAI_fraud\Trade_Map_-_List_of_importing_markets_for_the_product_exported_by_Morocco_in_2022.xls'  
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')

tables = pd.read_html(str(soup))
for i, table in enumerate(tables):

    output_directory = r'C:\Users\Admin\Desktop\MoroccoAI_fraud'
    os.makedirs(output_directory, exist_ok=True)
    csv_file_path = os.path.join(output_directory, f'table_{i}.csv')
    
    table.to_csv(csv_file_path, index=False)
```
